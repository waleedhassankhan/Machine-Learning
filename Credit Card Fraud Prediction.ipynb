# This data is highly unbalanced

O represent normal transection
1 represent fraud transection

```python
#seperating the data for analysis
legal = card_data[card_data.Class==0]
#if class value is O then entire row is stored in legal variable
fraud = card_data[card_data.Class==1]
#if class value is 1 then entire row is stored in fraud variable
```

# Under Sampling
Bulit a sample data containing the similar distribution of
normal tansectionand fraud transection

number of fraud transection is 492

```python
legal_sample = legal.sample(n=492)
```

```python
print(legal_sample.shape,fraud.shape)
```

```python
#concatenating two dataframe
new_data = pd.concat([legal_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

Splitting the data into target and features

```python
data = new_data.drop(columns='Class',axis=1)
label=new_data['Class']
```

```python
data.head()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(data,label,test_size=.2,stratify=label ,random_state=2)
```

```python
print(data.shape,X_train.shape,X_test.shape)
```

```python
model = LogisticRegression()
```

```python
#training the data
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)

print(test_data_accuracy)
```
