```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing data sets

```python
house_price_data = sklearn.datasets.load_boston()
```

```python
print(house_price_data)
```

```python
# Loading the datasets to a pandas data frame

house_price_dataframe = pd.DataFrame(house_price_data.data, columns=house_price_data.feature_names)
```

```python
house_price_dataframe.head()
```

```python
# add target (House price) column to the dataframe
house_price_dataframe['Price'] = house_price_data.target
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
#check for missing values
house_price_dataframe.isnull().sum()
```

```python
#statistical measures of the data
house_price_dataframe.describe()
```

understanding tthe coorelation between various features in a datasets

1. Positive Coorelation
2. Negative Coorelation

```python
coorelation = house_price_dataframe.corr()
#coorelation data stored in coorelation
```

```python
#constructing a heapmap to understand the coorelaton
plt.figure(figsize=(10,10))
sns.heatmap(coorelation, cbar= True, square=True, fmt= '.1f', annot=True, annot_kws= {'size':8}, cmap = 'Blues')
```

```python
#Spliting the data and the target(Price)

X_data = house_price_dataframe.drop(['Price'], axis = 1)
Y_prices = house_price_dataframe['Price']
```

```python
X_data
```

```python
Y_prices
```

```python
#splitting the data into test and train data
X_train, X_test, Y_train, Y_test = train_test_split(X_data,Y_prices, test_size = .2, random_state=2)
```

```python
print(X_data.shape,X_train.shape, X_test.shape)
```

```python
#Model Training
#Loading the model
model = XGBRegressor()
```

```python
#fitting/Training the data
model.fit(X_train,Y_train)
```

```python
#Evalution
#Predict the training data
#accuracy for  the prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
```

```python
print('R square error: ', score_1)
```

```python
print('Mean Absolute error: ', score_2)
```

# Visualize the actual prize and predicted price

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices va Predicted Prices')
```

```python
# Prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print('R square error: ', score_1)
print('Mean Absolute error: ', score_2)
```
